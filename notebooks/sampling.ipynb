# Random sampling

```python
SEED = 420
```

```python
from tally import Composition

for i in range(20):
    composition = Composition.random(hash((SEED, i)))
    composition.draw()
    label = input()
    print(f"Composition {i + 1} was given label {label}")
    composition.draw(path=f"data/{label}/{i + 1}.png")
    composition.save(path=f"data/{label}/{i + 1}.json")
```
